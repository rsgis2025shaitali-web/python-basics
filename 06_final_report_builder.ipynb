1. Vertical Structure and CompositionThe atmosphere is structured into layers defined by their temperature gradients, with the troposphere containing the bulk of aerosol mass and weather activity. Within this layer, atmospheric stability—determined by how temperature changes with height—dictates whether aerosols remain trapped near the surface or are mixed vertically.
2. Hydrostatic Balance and Pressure-Height RelationshipThe atmosphere exists in hydrostatic balance, where the downward force of gravity is perfectly countered by the upward pressure gradient force. This relationship explains why atmospheric pressure decreases monotonically with altitude; as you move higher, there is less overlying air mass to exert weight.
3. Concept of Scale HeightThe scale height ($H$) is the vertical distance over which atmospheric pressure or density decreases by a factor of $e$ (approx. 2.71). It serves as a measure of the atmosphere's "thickness" and is physically determined by the temperature and composition of the air.
