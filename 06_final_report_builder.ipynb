Write your summary below.
1. Coriolis Parameter and Latitude DependenceThe Coriolis parameter ($f$) represents the effect of Earth’s rotation on moving objects. It is highly dependent on latitude ($\phi$); it is zero at the equator and reaches its maximum at the poles. This variation means that the rotational "deflection" of winds and currents is much stronger at high latitudes than in the tropics.
2. Geostrophic Balance and Geostrophic WindGeostrophic balance occurs when the horizontal pressure gradient force (which pushes air from high to low pressure) is perfectly balanced by the Coriolis force (which pulls it to the right in the Northern Hemisphere). The resulting geostrophic wind flows parallel to isobars (lines of constant pressure) rather than across them. This balance governs the large-scale circulation of the atmosphere and oceans.
3. Rossby Number ($Ro$) and RotationThe Rossby number is a dimensionless ratio used to determine if Earth's rotation is a dominant factor in a fluid flow.Small $Ro$ ($Ro \ll 1$): Rotation is dominant, and the flow is nearly geostrophic (e.g., large-scale weather systems).Large $Ro$ ($Ro \gg 1$): Local forces (like friction or pressure) dominate over rotation (e.g., small-scale tornadoes or a draining bathtub).
4. Flow Examples and InterpretationsNorthern Hemisphere Flow: Winds circulate counter-clockwise around low-pressure systems (cyclonic) and clockwise around high-pressure systems (anticyclonic).Aerosol Transport: Large-scale geostrophic winds act as "conveyor belts" for aerosols. Once vertical instability lofts smoke or dust out of the boundary layer, these winds transport them thousands of kilometers across regions like the Indo-Gangetic Plain.

## 4. Summary of Notebook 3 - Thermodynamics & Stability (6–10 sentences)

Summarise the main thermodynamic concepts:
- Potential temperature and its interpretation  
- Environmental vs dry adiabatic lapse rate  
- Stability classification (stable, neutral, unstable)  
- Parcel method ideas and connection to convection

Highlight how these ideas help interpret the vertical structure of aerosols.

Write your summary below.
1. Potential Temperature ($\theta$) and InterpretationPotential temperature is the temperature an air parcel would have if it were moved adiabatically (without exchanging heat) to a reference pressure, typically 1000 hPa. Unlike regular temperature, $\theta$ stays constant during vertical movement if no heat is added or removed. It serves as a direct indicator of static stability: if $\theta$ increases with height, the atmosphere is stable; if it decreases, it is unstable.Environmental vs. Dry Adiabatic Lapse RateDry Adiabatic Lapse Rate ($\Gamma_d$): The constant rate at which a dry air parcel cools as it rises (approximately $9.8$ K/km).
2. Environmental Lapse Rate ($\Gamma$): The actual change in temperature with height measured by a sensor (e.g., a weather balloon). In the provided synthetic example, $\Gamma$ was estimated at $-9.04$ K/km, which is slightly less steep than the dry adiabatic rate.
3. Stability ClassificationThe relationship between the environmental lapse rate and the adiabatic rate determines the state of the atmosphere:Stable: The environment cools slower than the rising parcel. The parcel becomes colder and denser than its surroundings and sinks back down.Unstable: The environment cools faster than the parcel. The rising parcel remains warmer and less dense, causing it to accelerate upward.Neutral: The environmental lapse rate exactly equals the adiabatic lapse rate.
4. Parcel Method and ConvectionThe parcel method involves "nudging" a theoretical bubble of air upward to see if it continues to rise (convection) or returns to its original position. If the environment is unstable, convection occurs, creating a Deep Convective Boundary Layer that facilitates the vertical transport of heat, moisture, and particles.
5. Connection to Aerosol Vertical StructureThese thermodynamic principles dictate the "vertical volume" available for pollutants:Stable Conditions: Often found in the morning, stability acts as a "lid," trapping aerosols in a shallow layer near the surface, leading to higher observed AOD (e.g., 0.42).Unstable/Convective Conditions: Afternoon solar heating creates instability, allowing aerosols to be mixed high into the atmosphere. This "ventilation" dilutes surface concentrations and often reduces the total column AOD (e.g., 0.28) as particles are dispersed or exported by faster winds aloft.

Main Aerosol Types and Size ModesAerosols are categorized by their origin and size into two primary modes:Fine-mode aerosols: These are small particles (typically $< 1\mu m$) produced by combustion, such as urban pollution, industrial emissions, and biomass burning smoke.Coarse-mode aerosols: These are larger particles (typically $> 1\mu m$) resulting from mechanical processes, such as wind-blown desert dust and sea salt.
Key Optical DefinitionsAerosol Optical Depth (AOD): A dimensionless measure of how much light is attenuated (scattered or absorbed) by aerosols in a vertical column of the atmosphere.Single Scattering Albedo (SSA): The ratio of scattering to total extinction (scattering + absorption).Ångström Exponent ($\alpha$): A parameter that describes the wavelength dependence of AOD, used as an indicator of particle size.
Relating $\alpha$ to Particle SizeThe Ångström exponent is inversely related to particle size:High $\alpha$ (e.g., $1.5$ to $2.0$ or values closer to $0$ in negative synthetic sets): Indicates a dominance of small, fine-mode particles like smoke or urban pollution.Low $\alpha$ (e.g., $< 0.5$ or more negative values): Indicates a dominance of large, coarse-mode particles like dust or sea salt.
Relating SSA to Absorption vs. ScatteringThe SSA value indicates the "darkness" or reflectivity of an aerosol:Scattering Aerosols: Have an SSA close to 1.0 (e.g., Case 1: $0.960$). Examples include clean sea salt or sulfates.Absorbing Aerosols: Have lower SSA values (e.g., Case 3: $0.600$). Examples include black carbon (soot) or mineral dust, which warm the atmosphere by absorbing solar radiation.
Diurnal and Spectral PatternsDiurnal Cycle: In urban areas, AOD often peaks in the morning (0.42) due to traffic emissions trapped in a stable boundary layer, decreases in the afternoon (0.28) as convective mixing dilutes the column, and rebounds in the evening (0.35) as the atmosphere re-stabilizes.Spectral Pattern: AOD typically decreases as wavelength increases. The steepness of this decrease is what defines the Ångström exponent, with fine particles showing a much steeper spectral drop than coarse particles.
