### **Q2.2 — Classification**  
Classify each SST into:
- **HOT** (> 29)  
- **WARM** (28–29)  
- **NORMAL** (< 28)

Print classification for each day.  
At the end, print counts for each category.


````python
# Q2.2
# Define the list 'sst' with example values
sst = [27.5, 28.1, 29.5, 30.2, 28.9, 27.0, 31.1, 29.0]

for i, sst_value in enumerate(sst):
  if sst_value > 29:
    print(f"Day {i+1}: HOT")

  elif sst_value >= 28:
    print(f"Day {i+1}: WARM")
  else:
    print(f"Day {i+1}: NORMAL")

print(f"HOT: {len([1 for sst_value in sst if sst_value >29])}")

````

### **Q2.3 — Rolling 3-day Mean**  
Compute a 3-day rolling mean using loops (without NumPy).  
Example output:
```
Mean Day1-3: ...
Mean Day2-4: ...
...
```


````python
# Q2.3
# Define the list 'sst' with example values
sst = [27.5, 28.1, 29.5, 30.2, 28.9, 27.0, 31.1, 29.0]

# compute a 3-day rolling mean
for i in range (len(sst) -2):
  three_day_slice = sst[i:i+3]
  rolling_mean = sum(three_day_slice) / len(three_day_slice)
  print(f"Mean Day{i+1}-{i+3}: {rolling_mean}")
````
